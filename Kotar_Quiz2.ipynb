Q15. Create a new column called 'Category' on the tfl DataFrame that classifies the number of bike hires per day as:

    'Low' if the 'Number of Bicycle Hires' is below 10,000
    'Medium' if the 'Number of Bicycle Hires' is below 40,000 but greater than or equal to 10,000
    'High' if the 'Number of Bicycle Hires' is greater than or equal to 40,000

```python
conditions = [
    (tfl['Number of Bicycle Hires'] < 10000),
    (tfl['Number of Bicycle Hires'] >= 10000) & (tfl['Number of Bicycle Hires'] < 40000),
    (tfl['Number of Bicycle Hires'] >= 40000)
    ]
values = ['Low', 'Medium', 'High']
tfl['Category'] = np.select(conditions, values)
tfl
```

Q16. For each year in the tfl DataFrame how many days are classed as 'Low', 'Medium' or 'High'?

```python
tfl_groupby=tfl.groupby(['Year','Category']).count()
#tfl_groupby.drop("Number of Bicycle Hires")
tfl_groupby.drop('Number of Bicycle Hires', axis=1, inplace=True)
tfl_groupby

```
