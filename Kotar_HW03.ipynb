```python
from matplotlib import pyplot as plt, patches
class rectangle:
    def __init__(self,a,b,x1,y1):
        self.a=a
        self.b=b
        self.x1=x1
        self.y1=y1 
        self.length=self.x1+self.b
        self.width=self.y1+self.a
        self.origin=(self.x1,self.y1)
        
        
        
    def area(self):
            return self.a*self.b
    
    def perimeter(self):
        return 2*(self.a+self.b)
    
    def plot_shape(self):
        
        cord = (self.x1,self.y1)
        rec = plt.Rectangle((cord),width=self.a,height=self.b,edgecolor='blue',fc='white')

        ax=plt.gca()
        ax.add_patch(rec)
        plt.axis("scaled")
        return plt.show()

    def shift(self,delta_x,delta_y):
        self.delta_x=delta_x
        self.delta_y=delta_y
        self.x1 = self.x1+self.delta_x
        self.y1 = self.y1+self.delta_y
        return (self.x1,self.y1)
    
    
        
    
```

```python
# create a new rectangle object
r1 = rectangle(5,3,0,0)
```

```python
# get the area
r1.area()
```

```python
# get the perimeter
r1.perimeter()
```

```python
# plot the rectangle
r1.plot_shape()
```

```python
# shift the rectangle
r1.shift(2,2)
```

```python
# plot it and check the shift
r1.plot_shape()
```
